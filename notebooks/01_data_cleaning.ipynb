## About...

This project is an interactive machine learning simulator that helps students and educators understand how factors like study time, attendance, and learning environment affect academic performance. Users can adjust inputs and instantly see how predicted scores change.

#### Target
final score

#### Features
study_hours,
attendance_rate,
past_exam_scores,
parental_education,
internet_access,
extracurricular_activities

```python
import pandas as pd
import numpy as np 

# Load dataset
df = pd.read_csv("../data/raw/student_performance_dataset.csv")

# Preview
df.head()
```

```python
df.info()
df.describe()
```

```python
columns_needed = [
    "Study_Hours_per_Week",
    "Attendance_Rate",
    "Past_Exam_Scores",
    "Parental_Education_Level",
    "Internet_Access_at_Home",
    "Extracurricular_Activities",
    "Final_Exam_Score"
]

df = df[columns_needed]
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df["Internet_Access_at_Home"] = df["Internet_Access_at_Home"].map({"Yes": 1, "No": 0})

df["Extracurricular_Activities"] = df["Extracurricular_Activities"].map({"Yes": 1, "No": 0})
```

```python
df["Parental_Education_Level"].unique()
```

```python
df["Parental_Education_Level"] = df["Parental_Education_Level"].map({
    "High School": 0,
    "Bachelors": 1,
    "Masters": 2,
    "PhD":3
})
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df = df[df["Attendance_Rate"] <= 100]
```

```python
df.to_csv("../data/processed/cleaned_student_data.csv", index=False)
```
